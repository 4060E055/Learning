```python
import numpy as np
import cv2

cap = cv2.VideoCapture('./video/rabbit.mp4')   # play video file
# cap = cv2.VideoCapture(0)                         # from camera
fps = cap.get(cv2.CAP_PROP_FPS)               # Frame Per Second
F_Count = cap.get(cv2.CAP_PROP_FRAME_COUNT)   # frame count
size_h=cap.get(cv2.CAP_PROP_FRAME_WIDTH )
size_w=cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
print("size:",size_h,size_w)
print(f'fps : {fps:.2f} ms, Frame_Count : {F_Count}')
n=0
fgbg = cv2.bgsegm.createBackgroundSubtractorMOG(history=1, nmixtures=5, backgroundRatio=0.7)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(r'C:\Users\Tibame_T14\我的雲端硬碟\OpenCV\video\HW.mp4', fourcc, fps,(int(size_h),int(size_w)))


while cap.isOpened():
    ret, frame = cap.read()

    if not ret : break
    n+=1
    
    if n<50:
        frame=cv2.putText(frame, "Original image", (10, 900), cv2.FONT_HERSHEY_SIMPLEX, 1, (200, 100, 205), 2, cv2.LINE_AA)
                   
    elif 50<=n<120:
        frame=cv2.add(frame,frame.copy())
        frame=cv2.putText(frame, "Over exposure", (10, 900), cv2.FONT_HERSHEY_SIMPLEX, 1,  (200, 100, 205), 2, cv2.LINE_AA)
        
    elif 120<=n<190:
        frame=cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame=cv2.putText(frame, "BGR To RGB", (10, 900), cv2.FONT_HERSHEY_SIMPLEX, 1,  (200, 100, 205), 2, cv2.LINE_AA)
        
    elif 190<=n<260:
        frame=cv2.boxFilter(frame, -1, (10, 10), normalize=1) 
        frame=cv2.putText(frame, "Vague", (10, 900), cv2.FONT_HERSHEY_SIMPLEX, 1,  (200, 100, 205), 2, cv2.LINE_AA)
        
    elif 260<=n<340:
        sobelx = cv2.Sobel(frame, cv2.CV_64F, 1, 0, ksize=3)
        sobely = cv2.Sobel(frame, cv2.CV_64F, 0, 1, ksize=3)
        sobelx = cv2.convertScaleAbs(sobelx)   # 轉回 Uint8 
        sobely = cv2.convertScaleAbs(sobely)  
        frame =  cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0) 
        frame=cv2.putText(frame, "Sobel", (10, 900), cv2.FONT_HERSHEY_SIMPLEX, 1, (200, 100, 205), 2, cv2.LINE_AA)
        
    elif 340<=n<410:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        frame=cv2.putText(frame, "Gray", (10, 900), cv2.FONT_HERSHEY_SIMPLEX, 1,  (200, 100, 205), 2, cv2.LINE_AA)
   
    elif 410<=n<480:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        frame=cv2.equalizeHist(frame)
        frame=cv2.putText(frame, "Gray level Histogram Equalization", (10, 900), cv2.FONT_HERSHEY_SIMPLEX, 1, (200, 100, 205), 2, cv2.LINE_AA)
        
    else: 
        frame = cv2.flip(frame, 1)   # left side right
        frame=cv2.putText(frame, "Flip left and right", (10, 900), cv2.FONT_HERSHEY_SIMPLEX, 1,  (200, 100, 205), 2, cv2.LINE_AA)
        
        
    c=cv2.waitKey(30)
    if c==27:          # key escape
        break
    
    out.write(frame)
    cv2.imshow('frame',frame)

cap.release()
out.release()

cv2.destroyAllWindows()
cv2.waitKey(1)
```
